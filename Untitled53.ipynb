```python
import tkinter as tk
import time

class StopwatchApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Stopwatch")

        self.is_running = False
        self.start_time = 0

        self.minutes = 0
        self.seconds = 0
        self.milliseconds = 0

        self.time_label = tk.Label(root, text="00:00.000", font=("Arial", 24))
        self.time_label.pack(pady=20)

        self.start_button = tk.Button(root, text="Start", width=10, command=self.start_timer)
        self.pause_button = tk.Button(root, text="Pause", width=10, command=self.pause_timer)
        self.reset_button = tk.Button(root, text="Reset", width=10, command=self.reset_timer)

        self.start_button.pack()
        self.pause_button.pack()
        self.reset_button.pack()

        self.update_time()

    def start_timer(self):
        if not self.is_running:
            self.is_running = True
            self.start_time = time.time() - (self.minutes * 60 + self.seconds + self.milliseconds / 1000)
            self.update_timer()

    def pause_timer(self):
        if self.is_running:
            self.is_running = False

    def reset_timer(self):
        self.is_running = False
        self.minutes = 0
        self.seconds = 0
        self.milliseconds = 0
        self.update_time()

    def update_timer(self):
        if self.is_running:
            current_time = time.time()
            elapsed_time = current_time - self.start_time
            self.minutes, self.seconds = divmod(int(elapsed_time), 60)
            self.milliseconds = int((elapsed_time - int(elapsed_time)) * 1000)
            self.update_time()
            self.root.after(1, self.update_timer)

    def update_time(self):
        time_str = f"{self.minutes:02d}:{self.seconds:02d}.{self.milliseconds:03d}"
        self.time_label.config(text=time_str)

if __name__ == "__main__":
    root = tk.Tk()
    app = StopwatchApp(root)
    root.mainloop()
```

```python
import tkinter as tk
import time

class StopwatchApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Stopwatch")

        self.is_running = False
        self.start_time = 0
        self.elapsed_time = 0

        self.time_label = tk.Label(root, text="00:00:00", font=("Helvetica", 48))
        self.start_button = tk.Button(root, text="Start", command=self.start_stop)
        self.reset_button = tk.Button(root, text="Reset", command=self.reset)

        self.time_label.pack(pady=20)
        self.start_button.pack(pady=10)
        self.reset_button.pack()

        self.update_time()

    def start_stop(self):
        if self.is_running:
            self.is_running = False
            self.start_button.config(text="Start")
        else:
            self.is_running = True
            self.start_button.config(text="Pause")
            self.start_time = time.time() - self.elapsed_time
            self.update_time()

    def reset(self):
        self.is_running = False
        self.start_button.config(text="Start")
        self.start_time = 0
        self.elapsed_time = 0
        self.update_time()

    def update_time(self):
        if self.is_running:
            self.elapsed_time = time.time() - self.start_time

        minutes, seconds = divmod(int(self.elapsed_time), 60)
        milliseconds = int((self.elapsed_time - int(self.elapsed_time)) * 1000)
        self.time_label.config(text=f"{minutes:02d}:{seconds:02d}:{milliseconds:03d}")
        if self.is_running:
            self.root.after(10, self.update_time)

if __name__ == "__main__":
    root = tk.Tk()
    app = StopwatchApp(root)
    root.mainloop()
```
